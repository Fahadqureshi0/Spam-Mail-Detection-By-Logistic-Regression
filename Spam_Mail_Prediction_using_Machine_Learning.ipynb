```python
# Importing Depedencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

**Data Collection and Pre-Processing**

```python
# Loading the Data from csv to pandas Dataframes:
raw_mail_data = pd.read_csv('/content/mail_data.csv')
```

```python
print(raw_mail_data)
```

```python
# Replacing null values with empty strings:
mail_data = raw_mail_data.where((pd.notnull(raw_mail_data)),'')
```

```python
# Printing the 5 rows of Dataset:
mail_data.head()
```

```python
# Checking the number of rows and columns in Dataset:
mail_data.shape
```

**Label Encoding**

```python
# Label Spam mail as -->0
# ham mail as -->1
```

```python
mail_data.loc[mail_data['Category'] == 'spam' , 'Category',] = 0
mail_data.loc[mail_data['Category'] == 'ham' , 'Category',] = 1
```

Spam ==> 1
Ham ==> 0


```python
# Seprating the Data as texts and labels:
X = mail_data['Message']

Y = mail_data['Category']
```

```python
print(X)
```

```python
print(Y)
```

**Splitting the Data into Tarinig and Test Data**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=3)
```

```python
print(X.shape) # Origional Data
print(X_train.shape) # Training Data
print(X_test.shape) # Test Data
```

**Feature Extraction**

```python
from sklearn import feature_extraction
# Transforming the text data into feature vectors that can be used as input Logistic Regression Model
feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

# Converting Y_train and Y_test values into Integers:
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_train)
```

```python
print(X_train_features)
```

**Training the Model**

**Logistic Regression**

```python
model = LogisticRegression()
```

```python
# Training the LogisticRegression with the training data
model = model.fit(X_train_features, Y_train)
```

**Evaluating Trained Model**

```python
# Prediction on Training Model
prediction_on_training_data = model.predict(X_train_features)
accuracy_score_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
```

```python
print('Accuracy on Training Data: ', accuracy_score_on_training_data)
```

```python
# Prediction on testing Model
prediction_on_testing_data = model.predict(X_test_features)
accuracy_score_on_testing_data = accuracy_score(Y_test, prediction_on_testing_data)
```

```python
print('Accuracy on Testing Data: ', accuracy_score_on_testing_data)
```

**Predictive System**

```python
input_mail = ["WINNER!! As a valued network customer you have been selected to receivea £900 prize reward! To claim call 09061701461. Claim code KL341. Valid 12 hours only"]
# Convert Text to Feature Vectors:
input_data_features = feature_extraction.transform(input_mail)

# Making Prediction:
prediction = model.predict(input_data_features)
print(prediction)

if (prediction[0]==1):
  print('Ham Mail')
else:
  print('Spam Mail')
```
